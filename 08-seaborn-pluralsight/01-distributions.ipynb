```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style="darkgrid")
```

## 01 Getting Wine Data

```python
filename = 'course-files/02/demos/datasets/winequality-white.csv'

names=["Fixed Acidity", "Volatile Acidity", "Citric Acid", "Residual Sugar",
       "Chlorides", "Free Sulfur Dioxide", "Total Sulfur Dioxide", "Density", 
       "pH", "Sulphates", "Alcohol", "Quality"]
```

```python
wine_data = pd.read_csv(filename, sep=';')
```

```python
wine_data.head()
```

Let's now use the columns from `names` array.

```python
wine_data = pd.read_csv(filename, sep=';', names=names, skiprows=1)
```

```python
wine_data.head()
```

## 02 Distribution Plotting

### Different functions

```python
plt.figure(figsize=(5, 5))
sns.distplot(wine_data['Alcohol'])
```

```python
plt.figure(figsize=(5, 5));
sns.displot(wine_data['Alcohol'], kde=True);
```

```python
plt.figure(figsize=(5, 5));
sns.histplot(wine_data['Alcohol'], kde=True);
```

```python
plt.figure(figsize=(5, 5));
sns.kdeplot(data=wine_data['Alcohol'], fill=True);
```

### Rug plot

```python
_, ax = plt.subplots(figsize=(10, 3));
sns.distplot(wine_data['Alcohol'], kde=True, rug=True, ax=ax);
plt.tight_layout();
```

```python
_, ax = plt.subplots(figsize=(10, 3));
sns.distplot(wine_data['Alcohol'], kde=True, rug=True, ax=ax, bins=75);
plt.tight_layout();
```

```python
_, ax = plt.subplots(figsize=(10, 3));
sns.rugplot(wine_data['Alcohol'], height=.5, ax=ax);
plt.tight_layout();
```

### Different parameters of `kdeplot`

```python
f,ax=plt.subplots(figsize=(15,5))
sns.kdeplot(wine_data['Alcohol'])
sns.kdeplot(wine_data['Alcohol'], bw_method=.04, label="bw: 0.04")
sns.kdeplot(wine_data['Alcohol'], bw_method=.2, label="bw: 0.2")
sns.kdeplot(wine_data['Alcohol'], bw_method=2, label="bw: 2")
sns.kdeplot(wine_data['Alcohol'], bw_method=5, label="bw: 5")
plt.tight_layout();
```

### Heatmap

```python
corrmat = wine_data.corr()
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corrmat, vmax=.8, square=True, annot=True, fmt='.2f', cmap = "summer");
```

### Regression: simple regression

```python
plt.figure(figsize=(5, 3))
sns.regplot(x='Residual Sugar', y='Density', 
            data=wine_data, line_kws={'color': 'red'});
```

```python
# Drop outliers from the dataset where Residual Sugar is greater than 30
wine_data = wine_data[wine_data['Residual Sugar'] < 20]
```

```python
plt.figure(figsize=(15, 6))
sns.regplot(x='Residual Sugar', y='Density', ci=99,
            data=wine_data, line_kws={'color': 'red'},
            scatter_kws={'alpha': 0.4});
```

```python
wine_data_sample = wine_data.sample(100)
plt.figure(figsize=(15, 6))
sns.regplot(x='Residual Sugar', y='Density', ci=95,
            data=wine_data_sample, line_kws={'color': 'red'},
            scatter_kws={'alpha': 0.4});
```

### Regression: qualitative variables

```python
plt.figure(figsize=(5, 3))
sns.regplot(x='Quality', y='Alcohol', data=wine_data,
            line_kws={'color': 'red'},
            scatter_kws={'alpha': 0.4});
```

```python
plt.figure(figsize=(5, 3))
sns.regplot(x='Quality', y='Alcohol', data=wine_data,
            line_kws={'color': 'red'},
            scatter_kws={'alpha': 0.9},
            x_estimator=np.mean);
```

```python
plt.figure(figsize=(5, 3))
sns.regplot(x='pH', y='Fixed Acidity', data=wine_data,
            line_kws={'color': 'red'},
            scatter_kws={'alpha': 0.4});
```
