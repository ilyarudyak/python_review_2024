```python
# Generate random data for the histogram
data = np.random.randn(1000)

# Get the available styles
styles = plt.style.available

# Create a figure for each style
for style in styles:
    with plt.style.context(style):
        plt.figure(figsize=(4, 4))  # Create a new figure for each style
        plt.hist(data, bins=30, alpha=0.7)  # Create histogram
        plt.title(f'Histogram with style: {style}', fontsize=10)  # Set title
        plt.xlabel('Value')  # Set x-label
        plt.ylabel('Frequency')  # Set y-label
```
